```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import platform
```

```python
df = pd.read_csv('final_sleep_analysis.csv')
```

```python
features = ['Core', 'Deep', 'REM', 'Awake', 'InBed', 'TotalSleep', 'BedTime_Offset']
x = df.loc[:, features].values
y = df.loc[:, 'Date'].values
```

```python
x.size
```

```python
x = StandardScaler().fit_transform(x)
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data=principalComponents, 
                           columns=['PC1', 'PC2'])
```

```python
print("\n--- 目的1：主成分の解釈（因子負荷量） ---")
loadings = pd.DataFrame(pca.components_.T, columns=['PC1', 'PC2'], index=features)
print(loadings)
print("-" * 30)
```

```python
print("--- 目的1：次元削減の効果 ---")
print(f"各主成分の寄与率（Explained Variance Ratio）: {pca.explained_variance_ratio_}")
print(f"累積寄与率: {sum(pca.explained_variance_ratio_):.2f}")
```

```python
print("--- 目的2：Biplot の可視化 ---")

system_name = platform.system()
plt.rcParams['font.family'] = 'Hiragino Sans'

plt.figure(figsize=(12, 10), dpi=300)

plt.scatter(
    principalDf['PC1'],
    principalDf['PC2'],
    alpha=0.5,
    s=60,
    c='cornflowerblue',
    edgecolors='white',
    linewidth=0.5
)

scale_factor = 4.0

for i, feature in enumerate(features):
    plt.arrow(
        0,
        0,
        pca.components_[0, i] * scale_factor,
        pca.components_[1, i] * scale_factor,
        color='firebrick',
        alpha=0.9,
        width=0.03,
        head_width=0.15
    )

    plt.text(
        pca.components_[0, i] * scale_factor * 1.15,
        pca.components_[1, i] * scale_factor * 1.15,
        feature,
        color='black',
        fontsize=14,
        fontweight='bold',
        ha='center',
        va='center',
        bbox=dict(
            facecolor='white',
            alpha=0.7,
            edgecolor='none',
            boxstyle='round,pad=0.2'
        )
    )

plt.xlabel(
    f'第1主成分 (PC1) − 寄与率: {pca.explained_variance_ratio_[0] * 100:.1f}%',
    fontsize=14
)
plt.ylabel(
    f'第2主成分 (PC2) − 寄与率: {pca.explained_variance_ratio_[1] * 100:.1f}%',
    fontsize=14
)

plt.grid(True, linestyle='--', alpha=0.6)
plt.axhline(0, color='black', lw=1.2, linestyle='-')
plt.axvline(0, color='black', lw=1.2, linestyle='-')
# plt.savefig('pca_result_biplot_hd.png', dpi=300, bbox_inches='tight')
plt.show()
```
